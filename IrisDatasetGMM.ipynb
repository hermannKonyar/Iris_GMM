```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler

# Seaborn ktphanesinden iris veri kmesini ykleme
iris = sns.load_dataset('iris')

# Veriyi leklendirme
scaler = StandardScaler()
scaled_data = scaler.fit_transform(iris.drop(columns='species'))

# GMM modeli oluturma ve eitme
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(scaled_data)

# Snflandrma tahminleri
predictions = gmm.predict(scaled_data)

# Tahminleri DataFrame'e ekleme
iris['cluster'] = predictions

# Scatter plot ile grselletirme
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='cluster', palette='viridis')
plt.title('Iris Veri Kmesi GMM Snflandrmas')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Cluster')
plt.show()
```
